# importing libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Uploading image
[photo link](https://i.pinimg.com/236x/2d/e7/e8/2de7e89da9227c32ec3ecb3b235cdecc.jpg)

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Replace filename if needed
img = cv2.imread(list(uploaded.keys())[0])

# Convert BGR â†’ RGB (IMPORTANT for matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")

plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

In the following

```python
colors = ('r', 'g', 'b')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=col)

plt.title("RGB Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
resized = cv2.resize(img_rgb, (200, 200))

plt.imshow(resized)
plt.title("Resized Image (200x200)")
plt.axis("off")
```

```python
# Crop region: [y1:y2, x1:x2]
cropped = img_rgb[50:250, 50:250]

plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")
```

# Splitting three RGB channels

```python
R, G, B = cv2.split(img_rgb)


plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(R, cmap='gray')
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(G, cmap='gray')
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(B, cmap='gray')
plt.title("Blue Channel")
plt.axis("off")



zeros = np.zeros_like(R)

red_img   = cv2.merge([R, zeros, zeros])
green_img = cv2.merge([zeros, G, zeros])
blue_img  = cv2.merge([zeros, zeros, B])

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(red_img)
plt.title("Red Channel (Color)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(green_img)
plt.title("Green Channel (Color)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(blue_img)
plt.title("Blue Channel (Color)")
plt.axis("off")

plt.show()
```
